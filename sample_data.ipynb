# Write sampled data

```python
sample_rate = 0.025
```

```python
num_dialogs = len(data['data']['dialogs'])
```

```python
data['data']['dialogs'] = data['data']['dialogs'][0:int(num_dialogs*sample_rate)]
```

```python
train_f = 'data/visdial_1.0_train_sampled.json'
```

```python
with open(train_f, 'w') as outfile:
    json.dump(data, outfile)
```
